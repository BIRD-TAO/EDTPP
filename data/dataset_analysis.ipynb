```python
import pickle
file = '/usr/commondata/public/point_process_lht/edtpp_TKDE/data/stackoverflow/train_manifold_format.pkl'
with open(file, 'rb') as f:
    data = pickle.load(f)
```

```python
first_event = [i[0] for i in data['timestamps']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.hist(first_event, bins=30)
plt.ylabel('counts')
plt.xlabel('t')
plt.savefig('statis_on_sof.png')
```
